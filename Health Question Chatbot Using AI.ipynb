```python
from transformers import pipeline
```

```python
# chatbot pipeline
chatbot = pipeline(
    "text-generation",
    model="mistralai/Mistral-7B-Instruct-v0.1"
)
```

```python
question = input("Ask your health question: ")
```

```python
prompt = f"""
Act like a friendly medical assistant.

Rules:
- Give simple and safe answers
- Do not diagnose serious diseases
- Do not prescribe dangerous medicines
- Suggest visiting a doctor for emergencies

Question:
{question}
"""
```

```python
response = chatbot(
    prompt,
    max_new_tokens=150
)
print(response[0]["generated_text"])
```
