```python
import datasets

print(datasets.__version__)
```

```python
from datasets import load_dataset

dataset = load_dataset("empathetic_dialogues")
```

```python
from transformers import AutoTokenizer, AutoModelForCausalLM

# small chatbot model
model_name = "distilgpt2"

# tokenizer load
tokenizer = AutoTokenizer.from_pretrained(model_name)

# model load
model = AutoModelForCausalLM.from_pretrained(model_name)

# padding token set
tokenizer.pad_token = tokenizer.eos_token
```

```python
print(dataset)
```

```python
print(dataset["train"][0])
```

```python
def preprocess(example):

    text = (
        "User: " + str(example["utterance"]) +
        "\nBot: " + str(example["context"])
    )

    encoding = tokenizer(
        text,
        truncation=True,
        padding="max_length",
        max_length=128
    )

    # labels for training
    encoding["labels"] = encoding["input_ids"]

    return encoding
```

```python
tokenized_dataset = dataset["train"].map(preprocess)
```

```python
print(tokenized_dataset[0])
```

```python
small_dataset = tokenized_dataset.select(range(100))
```

```python
small_dataset.set_format(
    type="torch",
    columns=["input_ids", "attention_mask", "labels"]
)
```

```python
from transformers import TrainingArguments

training_args = TrainingArguments(

    output_dir="./results",

    per_device_train_batch_size=2,

    num_train_epochs=3,

    logging_steps=10
)
```

```python
from transformers import Trainer

trainer = Trainer(

    model=model,

    args=training_args,

    train_dataset=small_dataset
)
```

```python
trainer.train()
```

```python
from transformers import pipeline, GenerationConfig

# chatbot pipeline
chatbot = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer
)

# generation settings
generation_config = GenerationConfig(

    max_new_tokens=50,

    do_sample=True,

    #temperature=0.7,

    pad_token_id=tokenizer.eos_token_id
)

# user message
question = "User: I feel stressed and anxious.\nBot:"

# generate response
response = chatbot(

    question,

    generation_config=generation_config
)

# print response
print(response[0]["generated_text"])
```
